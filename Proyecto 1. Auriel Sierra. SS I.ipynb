# Introducción

En el año 2021, se estima que el 24.4% de las muertes en Estados Unidos fueron causadas por enfermedades del corazón e infartos. Para ese mismo año, en Europa, el 39 y 45% de las muertes totales y el 32 y 30% de muertes prematuras en hombres y mujeres, respectivamente, se deben a
enfermedades cardiovasculares. En ambos casos, dicho tipo de enfermedades fueron la mayor causa de muerte en la población. (Xu et al., 2022) (Timmis et al., 2022).
Los datos anteriores muestran la importancia de mejorar los métodos de prevención y detección de enfermedades cardiovasculares. Para está última, uno de los principales métodos para monitorear el comportamiento del corazón es el electrocardiograma (ECG), el cual es un examen no invasivoque registra la actividad eléctrica en el corazón. Los patrones presentes en un ECG se asocian a unaetapa del ciclo cardíaco. Los principales componentes de un ECG de observan en la figura 1. [Fuente]: Fuente: S. Kaplan Berkaya et al. / Biomedical Signal Processing and Control 43 (2018) 216–235. (Rosiek &
Leksowski, 2016).

Mediante un análisis correcto de un ECG, se pueden detectar una gran variedad de condiciones cardíacos como: arritmias, ritmo cardíaco anormal; enfermedades en arterias coronarias, insuficiencia de flujo sanguíneo al corazón debido al bloqueo de las arterias coronarias; entre otras. (Martis
et al., 2014; Rosiek & Leksowski, 2016).

Un ECG brinda mucha información acerca del corazón, la cual debe ser analizada por un médico para realizar una detección temprana de alguna anomalía o síntoma de una posible enfermedad cardíaca. Además, el comportamiento del corazón posee una naturaleza compleja. Es debido a esta gran cantidad de información que muchas veces se dan procedimientos inapropiados en su análisis
o errores por parte de los médicos. Realizar un estudio de los resultados del electrocardiograma y determinar si el paciente es candidato a alguna condición cardíaca es un arduo proceso, para el cual existen lineamientos establecidos por parte de la Sociedad Europea de Cardiología y por el Colegio de Cardiología Americana. (Taylor & Yang, 2020).

Los avances en tecnología de los tiempos han permitido que se lleven a cabo muchos proyectos para poder realizar análisis de ECG por métodos computacionales. Conforme pasa el tiempo, más partes del proceso de análisis de un ECG, desde el preprocesamiento del registro crudo hasta la
extracción de características puntuales, se ha intentado automatizar. Estos procesos pueden involucrar desde realizar un algoritmo predeterminado para eliminar el ruido presente en las señales, hasta implementar redes neuronales que permitan realizar un procesamiento completo de la señal o un reconocimiento de patrones que permita identificar condiciones cardíacas específicas. (Kachuee et al., 2018; Kaplan et al., 2018; Martis et al., 2014).

La principal limitante para lograr un análisis automatizado de ECG es que cada persona y cada condición cardíaca presenta patrones ligeramente distintos, por lo que es difícil brindar parámetros preestablecido. Las redes neuronales son capaces de crear algoritmos complejos que realicen funciones específicas como estas; el único inconveniente es que se necesitan muchas muestras de electrocardiogramas para perfeccionar una red como estas. Es por esto que se han creado varias bases de datos abiertas con el objetivo de proporcionar las herramientas para que se realice un avance en conjunto. Cabe agregar que existen muchos enfoques diferentes en los métodos de análisis y clasificación de las señales, incluso algunos que no involucran redes neuronales. (Kaplan et al., 2018).

En el presente trabajo se pretende realizar una representación gráfica de una señal de audio generada por la actividad del corazón. En este trabajo se emplean archivos de audio obtenidos de la base de muestras de Lui, Springer, Moody, et al. (2016) en An open access database for the evaluation of heart sound algorithms. Los fonocardiogramas (FCG) son registros de los sonidos
y murmullos producidos por el corazón, estos pueden emplearse para el estudio de la actividad cardíaca y, al igual que un ECG, sus diferentes estados se pueden asociar a una parte específica del ciclo cardíaco. La posible información que puede ser extraída de un FCG es menor que la de un ECG, sin embargo, el sonido proporciona información de la actividad mecánica del corazón que puede facilitar la detección de irregularidades o de condiciones específicas. (Ismail et al., 2018; Liu et al., 2016).

El mayor reto al trabajar con FCG es que los registros poseen distorsiones y ruido del ambiente, lo que las convierte en señales muy complejas. A pesar de esto, estas señales presentan dos pulsos distintivos por cada ciclo cardíaco. La automatización de un proceso de análisis y clasificación de señales de FCG se compone de una detección de posición de los pulsos y una clasificación de estos. Los pulsos se clasifican en normales (S1 y S2), anormales (S3 y S4) y murmullos (ruido). (Ismail et al., 2018). Al igual que las señales de un ECG, su automatización presenta diferentes retos y áreas a mejorar, sin embargo, cualquier avance en esta automatización de señales puede permitir un mayor alcance en los esfuerzos por proteger la salud de los habitantes ante la gran problemática de enfermedades
cardiovasculares.

**Pregunta 1.** ¿Cuál es la mayor amplitud presente en los datos?

En la mayoría de las gráficas, la mayor amplitud se da en forma de pulsos distinguibles del resto de la señal que ocurren de forma períodica. Estos pulsos se presentan en parejas, con un primer pulso con mayor amplitud que el segundo en la mayoría de los casos. El primer pulso tiene una amplitud aproximada de entre 3000 ua y 30000 u. y el pulso secundario de entre 2000 ua a 20000 ua.

El pulso con mayor amplitud representa el pulso S1 y el pulso con menor amplitud representa el pulso S2. El puslo S1 ocurre cuando se da la contraccion ventricular, que es cuando se bombea sangre oxigenada a todas las partes del cuerpo y sangre no oxigenada a los pulmones. El pulso S2 se da cuando se cierran las válvulas aórtica y pulmonar y la sangre oxigenada y no oxigenada comienza a llegar a el ventriculo derecho e izquierdo respectivamente. (Liu et al., 2016).

**Pregunta 2.** ¿Cuál sería el resultado sonoro de las siguientes operaciones sobre las señales: escalamiento, desplazamiento, reflexión?

El sonido que escuchamos proviene de la variación de presión del aire, la cual es percibida por el tímpano. De modo que el sonido depende de dos variables independientes que son la presión y la frecuencia. El rango de sonidos perceptibles se encuentra entre los 20 y 1000 Hz y entre los 0 y 120 dB. (Rao, 2008)

En las gráficas de la señales obtenidas, el eje y representa la intensidad del sonido y el eje x el paso del tiempo. Las gráficas de los sonidos representan una curva simplificada de la intensidad del sonido, no muestran la verdadera frecuencia del sonido ya que esta es muy alta y solo se vería como un bloque sólido, en vez de una fluctuación perceptible. 

Un escalamiento se refiere a multiplicar toda la señal por una factor a. Si a es mayor a 1, la señal aumenta su amplitud. Si a es menor a 1, la señal disminuye su amplitud. Además, el escalamiento no cambia la frecuencia de la señal. De modo que un escalamiento resultaría en un sonido más alto
(de mayor intensidad) o un sonido más bajo (de menor intensidad) dependiendo si el factor a es mayor o menor a 1, respectivamente.

Un desplazamiento de una señal con respecto al tiempo, provocaría un adelanto o atraso de la misma. De modo que el resultado sonoro de una señal desplazada sería que esta incia antes o después. El sonido no cambia como tal, solo cambia en momento en el que se genera.

Una reflexión de una señal es reflejar esta con respecto a un eje. Si se da una reflexión con respecto al eje x, el sonido no cambiaría, ya que el sonido es una señal alterna y es practicamente simétrica con respecto al eje x, por lo que hacer esta reflexión no generaría cambios perceptibles.
```python
import numpy as np
import random as rd
import wave
import sys
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy.signal import butter, lfilter, filtfilt 
from statistics import stdev
sys.path.insert(1, r'./../functions') 
# commands to create high-resolution figures with large labels
%config InlineBackend.figure_formats = {'png', 'retina'} 
plt.rcParams['axes.labelsize'] = 16 # fontsize for figure labels
plt.rcParams['axes.titlesize'] = 18 # fontsize for figure titles
plt.rcParams['font.size'] = 14 # fontsize for figure numbers
plt.rcParams['lines.linewidth'] = 1.4 # line width for plotting

#Definición de la función que extra los datos del archivo de audio 
def ecg(variable):
    record = wave.open(variable, 'r') 
    numChannels = record.getnchannels() 
    numFrames = record.getnframes() 
    sampleRate = record.getframerate() 
    sampleWidth = record.getsampwidth()
    dstr = record.readframes(numFrames * numChannels)
    waveData = np.frombuffer(dstr, np.int16)
    timeECG = np.linspace(0, len(waveData)/sampleRate, num=len(waveData))
    return timeECG, waveData
 
#Obteniendo los datos de la onda específica
timeECG, waveData = ecg("a0004.wav") 

#Estableciendo los parametros para la graficación
t1 = 0.8
t2 = 8
# Graficando la señal 
plt.figure(figsize=(18,6))
plt.xlabel(r'Tiempo (s)')
plt.ylabel(r'Magnitud')
plt.plot(timeECG,waveData, 'b')
plt.xlim(t1,t2)
plt.title("Acercamiento de a0004")
plt.show()
```

### Detectar Picos R

Ahora se crea una función que encuenta los máximo locales de las secciones de la señal que son mayores a un umbral específico. 

Este umbral se puede modificar de tal forma que solo los pulsos S1 sean detectados.

```python
def detecta_maximos_locales(timeECG, waveData, threshold_ratio=0.8):
    # If not all the R peaks are detected, lower the threshold_ratio
    # If components that are not R peaks (like T waves) are detected, up the threshold_ratio

    if len(timeECG) != len(waveData): #Raises an error if two arrays have different lengths
        raise Exception("The two arrays have different lengths.")
    
    interval = max(waveData) 
    threshold = threshold_ratio*interval 
    maxima = []
    maxima_indices = []
    mxs_indices = []
    banner = False
    
    for i in range(0, len(waveData)):
            
        if waveData[i] >= threshold:#If a threshold value is surpassed,
            # the indices and values are saved 
            banner = True
            maxima_indices.append(i)
            maxima.append(waveData[i])
            
        elif banner == True and waveData[i] < threshold: #If the threshold value is crossed
            # the index of the maximum value in the original array is saved
            index_local_max = maxima.index(max(maxima))
            mxs_indices.append(maxima_indices[index_local_max])
            maxima = []
            maxima_indices = []
            banner = False     

    return mxs_indices

mxs_indices = detecta_maximos_locales(timeECG, waveData)

plt.figure(figsize=(18,6))
plt.xlabel(r'time (s)')
plt.ylabel(r'voltage ($\mu$V)')
plt.xlim(min(timeECG),max(timeECG))
plt.plot(timeECG, waveData)
plt.scatter(timeECG[mxs_indices], waveData[mxs_indices], color='r')
plt.show()
```

### Calculo de la frecuencia cardíaca 

Ahora que ya se determinaron los pulsos S1, se hace un cálculo de la frecuencia cardíaca. 

Primero, se determinan los latidos por minuto promedio y luego, se grafican los latidos por minuto en cada intervalo de la señal antes establecida, junto con su desviación estándar y su promedio. 

```python
mean_bpm = 60*(len(mxs_indices)/(timeECG[-1]-timeECG[0]))
print("Latidos por minuto:", mean_bpm)
```

```python
heart_rate = [(60/(timeECG[mxs_indices[i+1]]-timeECG[mxs_indices[i]])) \
              for i in range(0, len(mxs_indices)-1)]
time = timeECG[mxs_indices]
mean_vec = [mean_bpm for i in range(0, len(time)-1)]
std_vec_plus = [mean_bpm+stdev(heart_rate) for i in range(0, len(time)-1)]
std_vec_minus = [mean_bpm-stdev(heart_rate) for i in range(0, len(time)-1)]

plt.fill_between(time[1:], std_vec_minus, std_vec_plus, facecolor="red",
                 label = "Standard deviation", color='red', alpha=0.4)
plt.plot(time[1:], mean_vec, c="r", label = "Mean heart rate")
plt.plot(time[1:], heart_rate, markersize=5, marker = "o", label="Heart rate")
plt.ylim((0, 200))
plt.title("Frecuencia Cardíaca")
plt.xlabel(r'Numero de cíclo')
plt.ylabel(r'Latidos por minuto')
plt.legend()
plt.show()
```
